```python
import os
from matplotlib import  pyplot as plt
import cv2
import numpy as np
import pandas as pd

%matplotlib inline
```

```python
d = './val_result'
```

### Validation result

```python
for i in range(47, 50):
    
    print('epoch %d' % i)
    
    plt.figure(figsize=(15, 5))
    
    plt.subplot(131)
    plt.title('image')
    img = cv2.imread(os.path.join(d, 'input_rgb_epoch_%d.png' % i))
    plt.imshow(img)
    
    plt.subplot(132)
    plt.title('gt')
    gt = cv2.imread(os.path.join(d, 'gt_depth_epoch_%d.png' % i))
    plt.imshow(gt)
    
    plt.subplot(133)
    plt.title('pred')
    pred = cv2.imread(os.path.join(d, 'pred_depth_epoch_%d.png' % i))
    plt.imshow(pred)
    
    plt.show()
```

### Learning curve

```python
df = pd.read_csv('./loss.csv')

plt.plot(df['train_loss'], label='train loss')
plt.plot(df['val_loss'], label='val loss')

plt.title('Loss curves')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend()
plt.grid()
```
